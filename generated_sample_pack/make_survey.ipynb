```python
import pandas as pd
import numpy as np

df=pd.read_csv("model_predictions.csv",header=None)
df.columns=["name","fc","fc_cat","cnn_cat","env_and_freq_cat","average_cat","stack_size","iteration_num"]
df=df.sample(frac=1)
```

```python
#how often does a stacksize result in a generation?
df.groupby('stack_size').count()
```

```python
#what are the group counts?
df=df[df["average_cat"]==df["cnn_cat"]]
z=df.groupby('average_cat').count()

z["name"]/np.sum(z["name"])
print(np.sum(z["name"]))
```

```python
df1=df[df["stack_size"]==1]
df2=df[df["stack_size"]==2]
df4=df[df["stack_size"]==4]
```

```python
survey_rows=pd.concat([df1.groupby('cnn_cat').head(20),df2.groupby('cnn_cat').head(20),df4.groupby('cnn_cat').head(20)])
survey_rows
```

```python
from shutil import copyfile

for n in survey_rows["name"]:
    copyfile("survey/"+n, "./filtered_survey/"+n)

sr=survey_rows.sort_values('name', ascending=True)
sr.to_csv("survey_answers.csv",index=False)
```

```python
surv=survey_rows[["name"]].copy()
surv=surv.sort_values('name', ascending=True)
surv=surv.reset_index(drop=True)
surv["drum_category"]=""
surv.to_csv("survey.csv",index=False)
```
